# 01 LangChain Agent


## 환경


```python
from pathlib import Path
import os

from dotenv import load_dotenv

project_root = Path.cwd()
if not (project_root / ".env.example").exists():
  project_root = project_root.parent

load_dotenv(project_root / ".env")

if not os.getenv("OPENAI_API_KEY"):
  raise RuntimeError("OPENAI_API_KEY is required")

chat_model = os.getenv("OPENAI_CHAT_MODEL", "gpt-5.5")
chat_model
```

## Tool
- tool설명은 악분 블로그 참고: https://malwareanalysis.tistory.com/926#ai-%EC%97%90%EC%9D%B4%EC%A0%84%ED%8A%B8%EC%9D%98-%EB%8C%80%ED%91%9C-%EA%B8%B0%EB%8A%A5-tools

```python
from collections import Counter

from langchain_core.tools import tool

@tool
def count_items(items: list[str]) -> dict[str, int]:
  """Count duplicate strings in a list."""
  return dict(Counter(items))
```

## Agent


```python
from langchain.agents import create_agent
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model=chat_model)
agent = create_agent(
  model=llm,
  tools=[count_items],
  system_prompt="Use tools for deterministic calculations.",
  debug=True,
)
```

## 실행


```python
fruit_list = ["Apple", "Banana", "Apple", "Peaches"]

result = agent.invoke({
  "messages": [
    {
      "role": "user",
      "content": f"Count each fruit in this list: {fruit_list}",
    }
  ]
})

result["messages"][-1].content
```
